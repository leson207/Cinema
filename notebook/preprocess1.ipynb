```python
import numpy as np
import pandas as pd

from tqdm import tqdm
tqdm.pandas()
```

```python
# https://www.kaggle.com/datasets/carolzhangdc/imdb-5000-movie-dataset

df = pd.read_csv('movie_metadata.csv')
df= df.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
df
```

```python
def actors_list(row):
    L=[]
    for col in ['actor_1_name', 'actor_2_name', 'actor_3_name']:
        if pd.notnull(row[col]):
            L.append(row[col])

    return L

df['actors']=df.progress_apply(actors_list,axis=1)
```

```python
def director_list(x):
    L=[]
    if pd.notnull(x):
        L.append(x)
    return L

df['director']=df['director_name'].progress_apply(director_list)
```

```python
df['genres']=df['genres'].progress_apply(lambda x: x.split('|'))
```

```python
df['Title'] = df['movie_title'].str.lower()
```

```python
df['Title'] = df['Title'].apply(lambda x : x[:-1])
df['director']=df['director_name']
```

```python
df=df[['Title','genres','director','actors']]
```
