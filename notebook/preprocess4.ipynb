```python
import pandas as pd

file_names=['processed_tmdb.csv','processed_imdb.csv', 'processed_wiki.csv']
data = [pd.read_csv(file_name) for file_name in file_names]
df = pd.concat(data)
```

```python
df.loc[:,'Title']=df['Title'].str.lower()
df=df.drop_duplicates(subset=['Title'])
df
```

```python
df.to_csv('final.csv', index=False)
```
