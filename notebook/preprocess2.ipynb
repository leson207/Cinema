```python
import numpy as np
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
```

```python
# https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset
credits=pd.read_csv('credits.csv')
credits
```

```python
metadata=pd.read_csv('movies_metadata.csv', low_memory=False)
metadata
```

```python
metadata['release_date'] = pd.to_datetime(metadata['release_date'], errors='coerce')
metadata['year'] = metadata['release_date'].dt.year

metadata['year'].value_counts().sort_index()
```

```python
metadata = metadata.loc[metadata.year <= 2017,['genres','id','title','year']]
```

```python
metadata['id'] = metadata['id'].astype(int)
```

```python
df = pd.merge(metadata, credits, on='id')
df
```

```python
import ast

def refine(x):
    L = []
    x=ast.literal_eval(x)
    for i in x:
        if i['name']=='Science Fiction':
            L.append('Sci-Fi')
        else:
            L.append(i['name'])

    return L

df['genres'] = df['genres'].progress_apply(refine)
```

```python
def refine(x):
    L = []
    for i in ast.literal_eval(x):
        L.append(i['name'])
        if(len(L)>=3):
            break

    return L

df['cast'] = df['cast'].progress_apply(refine)
```

```python
def refine(x):
    L = []
    for i in ast.literal_eval(x):
        if i['job']=='Director':
            L.append(i['name'])
            break

    return L

df['crew'] = df['crew'].progress_apply(refine)
```

```python
df = df[~df.isin([[]]).any(axis=1)]
df
```

```python
df.loc[:,'title'] = df['title'].str.lower()
df.loc[:,'member'] = df.loc[:,'cast']+df.loc[:,'crew']
df=df.drop_duplicates(subset ="title", keep = 'last')
```

```python
df.loc[:,'Title']=df['title']
df.loc[:,'director']=df['crew']
df.loc[:,'actors']=df['cast']
df=df[['Title','genres','director','actors']]
```

```python
df.to_csv('processed_tmdb.csv',index=False)
```
